```python
import redis

# Cấu hình kết nối
# Host là IP Tailscale hoặc 'localhost' nếu chạy trên cùng máy
REDIS_HOST = '100.117.51.34' 
REDIS_PORT = 6379
REDIS_PASS = 'hung12345'

try:
    # Kết nối
    r = redis.Redis(
        host=REDIS_HOST, 
        port=REDIS_PORT, 
        password=REDIS_PASS, 
        decode_responses=True # Để nhận string thay vì bytes
    )
    
    # Ping thử
    if r.ping():
        print(f"✅ Kết nối Redis thành công tại {REDIS_HOST}:{REDIS_PORT}")
        
        # Test ghi/đọc Stream cơ bản
        stream_key = "test_stream"
        
        # 1. Đẩy 1 tin nhắn vào
        msg_id = r.xadd(stream_key, {"status": "testing", "message": "Hello ETL"})
        print(f"   -> Đã đẩy tin nhắn ID: {msg_id}")
        
        # 2. Đọc tin nhắn ra
        # xread trả về: [[stream_name, [(msg_id, data)]]]
        messages = r.xread({stream_key: "0-0"}, count=1)
        for stream, msgs in messages:
            for msg_id, data in msgs:
                print(f"   -> Đã đọc tin nhắn: {data}")
                
        # Dọn dẹp
        r.delete(stream_key)
        
except Exception as e:
    print(f"❌ Lỗi kết nối Redis: {e}")
```
